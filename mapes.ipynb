# Mapes

Notebook amb plots de dades __kml__. Les dades s'agafen directament del servidor de l'[ajuntament](http://opendata.sabadell.cat/ca/inici/fitxes-cataleg?cod=OD-,KML&title=Dades%20en%20format%20KML)

(How to use pykml with python3 [link](http://installfights.blogspot.com/2018/04/how-to-run-pykml-in-python3.html)
    

```python
%matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 10)
from urllib.request import urlopen
from pykml import parser
from lxml import etree
import numpy as np

import matplotlib
from matplotlib.patches import Circle, Wedge, Polygon
from matplotlib.collections import PatchCollection
import matplotlib.pyplot as plt
```

## Districtes

```python
url = 'http://ca.sabadell.cat/fitxes/dadesobertes/kml/districtes.kml'

fileobject = urlopen(url)
root = parser.parse(fileobject).getroot()

cmap = matplotlib.cm.get_cmap('jet')



placemarks = [] 
for place in root.Document.Folder.Placemark:
    points_str = place.Polygon.outerBoundaryIs.LinearRing.coordinates
    
    points_list = str(points_str).split()
    points = np.zeros((len(points_list),2))

    for idx, pl in enumerate(points_list):
        plist = pl.split(',')
        points[idx,0] = float(plist[0])
        points[idx,1] = float(plist[1])
        
    placemarks.append(points)


fig, ax = plt.subplots()
for idx, p in enumerate(placemarks):
    ax.plot(p[:,0], p[:,1], alpha=0.7, color = cmap(1.*idx/len(placemarks)),
        linewidth=1, solid_capstyle='round', zorder=2)
ax.set_title('Districtes')
plt.show()
```
